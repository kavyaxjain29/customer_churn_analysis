```python
import pandas as pd

df = pd.read_excel("Telco_customer_churn.xlsx")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#Checking for null values
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("-", "_")
)
```

```python
df.columns
```

```python
# Checking for duplicate Customers ID's

df['customerid'].duplicated().sum()
```

```python
# Dropping columns 'count', 'country', 'state', 'city', 'zip_code', 'lat_long', 'latitude', 'longitude' as they don't contribute to customer churn analysis

drop_cols = [
    'count',
    'country',
    'state',
    'city',
    'zip_code',
    'lat_long',
    'latitude',
    'longitude'
]

df.drop(columns=drop_cols, inplace=True)
```

```python
df.dtypes
```

```python
#Converting 'total_charges' to a numeric column

df['total_charges'] = pd.to_numeric(df['total_charges'], errors='coerce')
```

```python
#Count of missing values created

df['total_charges'].isnull().sum()
```

```python
# Remove rows with missing Total Charges (11 rows, ~0.16% of the dataset)

df = df.dropna(subset=['total_charges'])
```

```python
#Checking for unique values

for col in df.select_dtypes('object'):
    print(col)
    print(df[col].unique())
```

```python
df.to_csv("clean_telco_churn.csv", index=False)
```
